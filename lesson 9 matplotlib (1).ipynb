```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

```python
plt.plot(years,iran_pop)
plt.show()
```

```python
plt.scatter(years,iran_pop)
plt.show()
```

```python
plt.figure(figsize=(11,5.5), dpi=480)
plt.plot(years,iran_pop, color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('POPULATION RATIO WITH YEAR')
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['22M','29M','39M','56M','66M','75M','80M'])

plt.show()
```

```python
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
turkey_pop = [11.91, 34.51, 38.67, 46.23, 56.13, 81.57, 83.28]
```

```python
plt.subplot(1,2,1)
plt.plot(years, iran_pop)
plt.title('IR IRAN Population')

plt.subplot(1,2,2)
plt.plot(years,turkey_pop)
plt.title('TURKEY R. Population')

plt.show()
```
